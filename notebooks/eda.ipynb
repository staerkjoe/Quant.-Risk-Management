```python
# Distribution classes
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
from pathlib import Path

# Add project root to Python path (go up one level from notebooks/)
project_root = Path.cwd().parent
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

from loaders.dataloader import DataLoader
from main import load_config


# Set plotting style
plt.style.use('default')
sns.set_palette("Set2")
```

```python
config = load_config()

loader = DataLoader(config)
df = loader.load_and_prepare()
```

```python
# define the features
ordinal_features = config['DataConfig']['ordinal_features']
nominal_features = config['DataConfig']['nominal_features']
num_features = config['DataConfig']['num_features']
```

```python
df.head()
```

```python
# Check null values
df.info()
```

```python
sns.countplot(x="credit_risk", data=df)
plt.title("Target distribution (Credit Risk)")
plt.show()
```

```python
sns.countplot(x="foreign_worker", data=df)
plt.title("Foreign Worker Distribution")
plt.show()
```

```python
for col in num_features:
    plt.figure(figsize=(8, 5))
    sns.histplot(data=df, x=col, hue="credit_risk", kde=True, element="step")
    plt.title(f"{col} distribution by Credit Risk")
    plt.tight_layout()
    plt.show()
```

```python
for col in num_features:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x="credit_risk", y=col, data=df)
    plt.title(f"{col} by Credit Risk")
    plt.tight_layout()
    plt.show()
```

```python
for col in nominal_features:
    plt.figure(figsize=(8, 5))
    sns.countplot(x=col, hue="credit_risk", data=df)
    plt.title(f"{col} distribution by Credit Risk")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
for col in ordinal_features:
    plt.figure(figsize=(8, 5))
    sns.countplot(x=col, hue="credit_risk", data=df)
    plt.title(f"{col} distribution by Credit Risk")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[num_features + ["credit_risk"]].corr(), annot=True, cmap="coolwarm", center=0)
plt.title("Correlation heatmap (numeric features)")
plt.show()
```
